4. High Covariance in Reviews:

### Year-wise Analysis

```python
df.groupby('Year')[['User Rating', 'Reviews', 'Price']].agg({'max','min','mean'}).reset_index()
```

```python
year_wise_data= df.groupby('Year')[['User Rating', 'Reviews', 'Price']].mean()
year_wise_data.reset_index(inplace=True)
```

```python
sns.lineplot(x=year_wise_data.Year,y=year_wise_data['Price'])
plt.title('Year wise Average Price Analysis')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

### Conclusion: 
The variation in book prices could be due to multiple factors such as promotional strategies, changes in publishing costs, and market demand. The decrease in average prices towards the end of the decade might reflect market competition and the availability of more affordable reading options like e-books.

```python
sns.lineplot(x=year_wise_data.Year,y=year_wise_data['User Rating'])
plt.title('Year wise Average User Ratings Analysis')
plt.xlabel('Year')
plt.ylabel('Average User Ratings')
plt.show()
```

### Conclussion: 
The user ratings suggest a general trend of increasing customer satisfaction over the years, with books receiving higher average ratings as time progresses.

```python
sns.lineplot(x=year_wise_data.Year,y=year_wise_data['Reviews'])
plt.title('Year wise Average Reviews Analysis')
plt.xlabel('Year')
plt.ylabel('Average Reviews')
plt.show()
```

### Conclusion: 
The substantial increase in the number of reviews over the years indicates growing engagement and readership for the best selling books. This trend might be influenced by increasing internet usage and the popularity of online reviews.
